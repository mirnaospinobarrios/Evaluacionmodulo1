```python
class TiendaOnline:
    def __init__(self):
        self.inventario = []
        self.clientes = {}
        self.ventas_totales = 0.0      

    def agregar_producto(self, nombre, precio, cantidad):
        for producto in self.inventario:
            if producto['nombre'] == nombre:
                producto['cantidad'] += cantidad    
                   
                print("estoy en el if")
                return self.inventario
        else:
                # Si no existe, agrega un nuevo producto al inventario
                nuevo_producto = {'nombre': nombre, 'precio': precio, 'cantidad': cantidad}
                self.inventario.append(nuevo_producto)
                print("estoy en el else")
                print(self.inventario)
                return self.inventario           
                    
    def ver_inventario(self):
        for producto in self.inventario:
            for clave, valor in producto.items():
                 print(f"{clave}: {valor},", end=" ")
            print()
            
    def buscar_producto_por_nombre(self, nombre):
            
    # Itera a través del inventario 
        for producto in self.inventario:
            if producto["nombre"]==nombre:
                 print (f"Buscar producto por nombre {producto}")
                 
    def actualizar_stock(self, nombre, cantidad=None):
        for producto in self.inventario:
         # condicinal para saber si el producto existe   
            if producto['nombre'] == nombre:
                if cantidad is not None:
                      producto[2] = cantidad
            
                      print(f"Producto {nombre} actualizado con éxito: a {cantidad} unidades")
           
                return cantidad
            else:
                print(f"No se encontró el producto  {nombre}")
                return
    def eliminar_producto(self, nombre):
        for producto in self.inventario:        
            if producto['nombre'] == nombre:                                    
                self.inventario.remove(producto)                          
                print("Producto eliminado con éxito.")                 
                return self.inventario
            else:
                    print(f"No se encontró el producto  {nombre} en el inventario.")
                    break
    
    def calcular_valor_inventario(self):
        valor_total = 0      
        for producto in self.inventario:
            precio = producto["precio"]
            cantidad = producto["cantidad"]
            valor_producto = precio*cantidad
            valor_total += valor_producto
        print(f"el valor total de inventario es : {valor_total} euros")    
        return valor_total
    
    def realizar_compra(self):
        
        while  True:
                                    
            nombre = input(f"Del siguiente inventario, {self.inventario} ingrese el nombre del producto que desea comprar (o 'fin' para terminar su compra): ")

            if nombre.lower() == 'fin':
                break
            cantidad = int(input(f"Ingrese la cantidad de {nombre} a comprar: "))
            valor_total = 0
            for producto in self.inventario:
                if producto['nombre'] == nombre:
                    if producto['cantidad'] >= cantidad:
                        producto['cantidad'] -= cantidad
                        valor_producto = cantidad*producto["precio"]
                        valor_total += valor_producto                                             
                        print(f"Compra de {cantidad} unidades de {nombre} y con un valor de : {valor_producto} ha sido realizada con éxito.")
                    else:
                        print(f"No hay suficiente stock de {nombre}. Stock disponible: {producto['cantidad']}")
                    break
             
            else:
                print(f"No se encontró el producto {nombre} en el inventario.")
            
            print(f"valor total de la compra ¨{valor_total}")
        return self.inventario
        
                
        
tienda = TiendaOnline()

tienda.agregar_producto('Camisa', 20, 50)
tienda.agregar_producto('pantalon', 20, 40)
tienda.agregar_producto("zapatos", 50, 40)
tienda.ver_inventario()
tienda.buscar_producto_por_nombre("camisa")
tienda.actualizar_stock("Vaquero", 3)
tienda.actualizar_stock('Camisa', 20)
tienda.eliminar_producto("Camisa")
tienda.eliminar_producto("Camis")   
tienda.calcular_valor_inventario()
tienda.realizar_compra()
```
